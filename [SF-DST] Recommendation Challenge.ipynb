```python
import numpy as np
import pandas as pd
import scipy.sparse as sparse
from lightfm import LightFM
from lightfm.cross_validation import random_train_test_split
from lightfm.evaluation import precision_at_k, recall_at_k
import sklearn
```

```python
# Импорт для локального использования
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
submission = pd.read_csv('sample_submission.csv')
```

```python
#число потоков нашего процессора
NUM_THREADS = 8 

#число параметров вектора 
NUM_COMPONENTS = 30 

#число эпох обучения
NUM_EPOCHS = 30 
```

```python
#Создаём модель
model = LightFM(learning_rate=0.05, loss='logistic', no_components=NUM_COMPONENTS)
```

```python
#создаём разряженную матрицу для работы нашей библиотеке
ratings_coo = sparse.coo_matrix(
    (train['rating'].astype(int),
     (train['userid'], train['itemid'])))
```

```python
#Обучаем модель
model = model.fit(ratings_coo, epochs=NUM_EPOCHS, num_threads=NUM_THREADS)
```

```python
preds = model.predict(test.userid.values,
                      test.itemid.values)
```

```python
normalized_preds = (preds - preds.min())/(preds - preds.min()).max()
```

```python
submission['rating'] = normalized_preds
submission.to_csv('submission.csv', index=False)
submission.head(3)
```
